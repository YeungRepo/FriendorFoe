```python

Max_Time_Ind = len(SubSampleIndex); 
All_Strain_Keys = list(Num_to_Curve_Dict.keys());
All_Strain_Keys.sort();


PlateScores = np.zeros((8,12,3))
#kernel_ind = 0;
#Strain_Call = np.zeros((len(all_row_ind),len(all_col_ind))); 
#Strain_Call_Names = np.array(Strain_Call,dtype=object); 

for this_kernel_ind in Select_Kernels:
    for row_ind in range(0,len(all_row_ind)):
        for col_ind in range(0,len(all_col_ind)):
            this_well = all_row_ind[row_ind] + all_col_ind[col_ind];
            time_keys = list(feature_dict[this_well].keys());
            time_keys.sort()
            #L2Scores = np.ones((len(All_Strain_Keys),3))*0;

            kernel =  kernel_names[this_kernel_ind];
            if this_kernel_ind==0:
                ys = [feature_dict[this_well][elem][kernel] for elem in time_keys]
                ys = ys/np.max(ys);
            if this_kernel_ind ==1:
                ys = [feature_dict[this_well][elem][kernel] for elem in time_keys]
                ys = [np.max([elem-np.max(ys[0:1]),0]) for elem in ys]
                if np.max(ys)>0:
                    ys = ys/np.max(ys);
            if this_kernel_ind==2:
                ys = [feature_dict[this_well][elem][kernel] for elem in time_keys]
                ys = [np.max([elem-ys[0],0]) for elem in ys]
                if np.max(ys)>0:
                    ys = ys/np.max(ys)
            ys = savgol_filter(ys[0:Max_Time_Ind], Max_Time_Ind-1, PolyOrd) 
            
            Reference_Curve = Num_to_Curve_Dict[PathogenOfChoice][this_kernel_ind];
            SubSampleIndices = np.linspace(0,Max_Time_Ind*3,2)
            PlateScores[row_ind,col_ind,this_kernel_ind] = np.sum(np.array(Reference_Curve)[SubSampleIndex]-np.array(ys)[0:Max_Time_Ind])

```
